```python
import requests
import cv2
import numpy as np
from io import BytesIO
from PIL import Image
from matplotlib import pyplot as plt

# ---- CONFIG ----
ROBOT_IP = "169.254.19.251"

URL = f"http://{ROBOT_IP}:31950/camera/picture"
HEADERS = {"Opentrons-Version": "2"}



```

```python
# ---- CAPTURE IMAGE ----
try:
    print("[INFO] Requesting image from OT-2 camera...")
    resp = requests.post(URL, headers=HEADERS, timeout=10)
    resp.raise_for_status()

    img = Image.open(BytesIO(resp.content))
    img.save("deck_capture.jpg")
    print("[OK] Saved image as deck_capture.jpg")

except Exception as e:
    print("[ERROR] Failed to capture image:", e)
    raise SystemExit
```

```python
# ---- CONVERT TO NUMPY + COLOR DETECTION ----
img_cv = cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)

# Example: detect red objects
lower_red = np.array([0, 70, 50])      # adjust as needed
upper_red = np.array([10, 255, 255])
hsv = cv2.cvtColor(img_cv, cv2.COLOR_BGR2HSV)
mask = cv2.inRange(hsv, lower_red, upper_red)
result = cv2.bitwise_and(img_cv, img_cv, mask=mask)
```

```python
# ---- DISPLAY RESULTS ----
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Detected Red Areas")
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
